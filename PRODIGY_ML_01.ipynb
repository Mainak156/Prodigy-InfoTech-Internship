```python
import numpy as np
import pandas as pd
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')
```

```python
train_df = train_df[['GrLivArea', 'BedroomAbvGr', 'FullBath', 'BsmtFullBath', 'HalfBath', 'BsmtHalfBath', 'SalePrice']]
test_df = test_df[['GrLivArea' ,'BedroomAbvGr', 'FullBath', 'BsmtFullBath', 'HalfBath', 'BsmtHalfBath', 'SalePrice']]
```

```python
test_df.dropna(axis=0, inplace=True)
test_df.isnull().sum()
```

```python
X_train, X_test, y_train, y_test = train_df.drop('SalePrice', axis=1), test_df.drop('SalePrice', axis=1), train_df['SalePrice'], test_df['SalePrice']
```

```python
X_train.head()
```

```python
X_train['Bathrooms'] = X_train['FullBath'] + X_train['HalfBath'] + X_train['BsmtFullBath'] + X_train['BsmtHalfBath']
X_train.head()
```

```python
X_test['Bathrooms'] = X_train['FullBath'] + X_train['HalfBath'] + X_train['BsmtFullBath'] + X_train['BsmtHalfBath']
X_test.head()
```

```python
X_train.drop(columns=['FullBath', 'HalfBath', 'BsmtFullBath', 'BsmtHalfBath'], axis=1, inplace= True)
```

```python
X_test.drop(columns=['FullBath', 'HalfBath', 'BsmtFullBath', 'BsmtHalfBath'], axis=1, inplace= True)
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
y_pred=lr.predict(X_test)
```

```python
mean_squared_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```

```python
cv_scores = cross_val_score(lr, X_train, y_train, cv=5, scoring='r2')
print(f'Cross-validated R2 Score: {cv_scores.mean():.2f}')
```
