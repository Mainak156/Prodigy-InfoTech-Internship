```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('Mall_Customers.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.tail()
```

```python
data['Gender'].value_counts()
```

```python
data['Gender'] = data['Gender'].apply(lambda x: 1 if x=='Male' else 0)
```

```python
data['Age'].value_counts()
```

```python
# 0 -- 18 - 28
# 1 -- 29 - 38
# 2 -- 39 - 48
# 3 -- 49 - 60
# 4 -- 61 - 70
data['Age_Group'] = data['Age'].apply(
    lambda x: 0 if (18 <= x <= 28) else 
              (1 if (29 <= x <= 38) else 
              (2 if (39 <= x <= 48) else 
              (3 if (49 <= x <= 60) else 
              (4 if (61 <= x <= 70) else 5)))))
```

```python
data['Age_Group'].value_counts()
```

```python
data.drop('Age', axis=1, inplace= True)
```

```python
data.tail(10)
```

```python
X = data[['Gender', 'Annual Income (k$)', 'Spending Score (1-100)', 'Age_Group']].values
```

```python
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans.fit(X)
```

```python
data['Cluster'] = kmeans.labels_
data = data.sort_values(by='Cluster')
```

```python
data.head(10)
```

```python
data['Gender'] = data['Gender'].apply(lambda x: 'Male' if x==1 else 'Female')
data.head(6)
```

```python
data['Cluster'].value_counts()
```

```python
data.to_csv('Mall_Customers_Grouped.csv', index=False)
```
